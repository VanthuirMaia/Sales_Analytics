Tipos:

Numéricos (int64, float64): identificadores, valores monetários, quantidades, descontos e lucros.

Categóricos (object): cliente, produto, categoria, cidade, estado, etc.

Datas (object): Order Date e Ship Date — precisam ser convertidas para datetime.



```python
df['Order Date'] = pd.to_datetime(df['Order Date'], dayfirst=False)  # Converte a coluna de data para o tipo datetime
df['Ship Date'] = pd.to_datetime(df['Ship Date'], dayfirst=False)  # Converte a coluna de data de envio para o tipo datetime
```

```python
df['Ano'] = df['Order Date'].dt.year  # Extrai o ano da data do pedido
df['Mes'] = df['Order Date'].dt.month  # Extrai o mês da data do pedido
df['Trimestre'] = df['Order Date'].dt.quarter  # Extrai o trimestre da data do pedido
```

## Calcular métricas (soma de vendas, soma de lucro, quantidade de pedidos)

## Visualizar com gráficos

```python
vendas_ano = df.groupby('Ano')[['Sales', 'Profit']].sum().reset_index()  # Agrupa por ano e calcula a soma de vendas e lucro
print(vendas_ano)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))

# Barras de vendas e lucro
bars_sales = plt.bar(vendas_ano['Ano'], vendas_ano['Sales'], 
                     color='skyblue', label='Vendas')
bars_profit = plt.bar(vendas_ano['Ano'], vendas_ano['Profit'], 
                      color='orange', label='Lucro', alpha=0.7)

# Rótulos para vendas
for bar in bars_sales:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval, f'{yval:,.0f}',
             ha='center', va='bottom', fontsize=9)

# Rótulos para lucro
for bar in bars_profit:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval, f'{yval:,.0f}',
             ha='center', va='bottom', fontsize=9)

plt.title('Vendas e Lucro por Ano')
plt.xlabel('Ano')
plt.ylabel('Valores (USD)')
plt.legend()
plt.show()

```

# Análise de Vendas por categoria

```python
vendas_categoria = df.groupby('Category')[['Sales', 'Profit']].sum().reset_index()  # Agrupa por categoria e calcula a soma de vendas e lucro
print(vendas_categoria)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 5))
plt.bar(vendas_categoria['Category'], vendas_categoria['Sales'], color='lightgreen', label='Vendas')
plt.bar(vendas_categoria['Category'], vendas_categoria['Profit'], color='coral', label='Lucro', alpha=0.7)

#Rótulo para Vendas
for i, v in enumerate(vendas_categoria['Sales']):
    plt.text(i, v + 1000, f'{v:,.0f}', ha='center', va='bottom')
#Rótulo para Lucro
for i, v in enumerate(vendas_categoria['Profit']):
    plt.text(i, v + 1000, f'{v:,.0f}', ha='center', va='bottom')
    
plt.title('Vendas / Lucro por Categoria')
plt.xlabel('Categoria')
plt.ylabel('Valores (USD)')
plt.legend()
plt.show()
```

# Análise de Vendas por Região

```python
vendas_regiao = df.groupby('Region')[['Sales', 'Profit']].sum().reset_index()  # Agrupa por região e calcula a soma de vendas e lucro
print(vendas_regiao)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 5))
plt.bar(vendas_regiao['Region'], vendas_regiao['Sales'], color='lightblue', label='Vendas')
plt.bar(vendas_regiao['Region'], vendas_regiao['Profit'], color='lightcoral', label='Lucro', alpha=0.7)

# Rótulo para Vendas
for i, v in enumerate(vendas_regiao['Sales']):
    plt.text(i, v + 1000, f'{v:,.0f}', ha='center', va='bottom')

# Rótulo para Lucro
for i, v in enumerate(vendas_regiao['Profit']):
    plt.text(i, v + 1000, f'{v:,.0f}', ha='center', va='bottom')

plt.title('Vendas / Lucro por Região')
plt.xlabel('Região')
plt.ylabel('Valores (USD)')
plt.legend()
plt.show()
```
